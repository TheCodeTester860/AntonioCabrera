### Introduccion

#### Conceptos
##### En programación, la minería de datos se conceptualiza como la aplicación técnica de algoritmos para encontrar patrones. No es tanto una "ciencia" holística, sino un proceso computacional específico.

#####Definición Técnica: Es el subconjunto de la informática centrado en el descubrimiento automático de patrones (correlaciones, anomalías, clústeres) en grandes conjuntos de datos utilizando métodos de la inteligencia artificial, aprendizaje automático, estadística y sistemas de bases de datos.

#####Concepto en Programación:

#####El "Motor": Para un programador, la minería de datos es la implementación de la lógica algorítmica. Es escribir el código que ejecuta un algoritmo de Clustering (agrupamiento) o de Regresión sobre una matriz de datos limpia.

#####Enfoque: Se centra en el algoritmo y su eficiencia. ¿Cómo procesamos 1 millón de filas para encontrar fraudes?

#####Bibliotecas comunes: En Python, esto se traduce en usar librerías como Scikit-learn (para algoritmos) o Weka (en Java).

#####Analogía: Si los datos son una montaña de tierra, la minería de datos es el pico y la pala (la herramienta y la técnica) que usas para extraer las pepitas de oro.

####  Ciencia de datos

#####La ciencia de datos es un concepto más arquitectónico y multidisciplinario. En programación, se ve como el ciclo de vida completo del manejo de datos, donde la programación es el pegamento que une estadística, negocio y tecnología.

#####Definición General: Es un campo interdisciplinario que utiliza métodos científicos, procesos, algoritmos y sistemas para extraer conocimiento e insights de datos estructurados y no estructurados. Abarca desde la limpieza de datos hasta la visualización y la toma de decisiones.

#####Concepto en Programación:

#####El "Pipeline" (Tubería): Para un programador, hacer Data Science implica construir un flujo de trabajo (pipeline). Implica escribir scripts para extraer datos (SQL/Scraping), limpiarlos (Pandas), analizarlos (donde entra la minería de datos) y visualizarlos (Matplotlib/Tableau).

#####Enfoque: Se centra en la solución del problema de negocio. El código es un medio para responder una pregunta compleja.

#####Herramientas: Lenguajes como Python y R, entornos como Jupyter Notebooks, y plataformas de nube (AWS, Azure).

#####Analogía: La ciencia de datos es toda la operación geológica: desde decidir dónde excavar, gestionar la logística de la mina, usar el pico y la pala (minería de datos), hasta refinar el oro y venderlo en el mercado.

### Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

### Promedio (Average) y Promedio Ponderado


```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

```python
print(promedio)
print(promedio_ponderado)
```

### Mediana


```python
mediana = np.median(data)
```

### Varianza y Desviación Estándar


```python
varianza = np.var(data)
desviacion = np.std(data)
```

### Generación de Datos Aleatorios
